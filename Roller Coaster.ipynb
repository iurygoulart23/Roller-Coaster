```python
# 1 importing library's and dastaset's
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
from matplotlib import figure

golden_wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
golden_steel = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')

print(golden_wood.head())
# print(golden_steel.head())
print(golden_wood['Year of Rank'])
```

```python
# 2 Creating a function to plot rankings over time for 1 roller coaster
def plot_ranking_overtime(coaster_name, park_name, ranking):

  ranking = ranking[(ranking['Name'] == coaster_name) & (ranking['Park'] == park_name)]
  

  fig, ax = plt.subplots(figsize=(8, 6))
  ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
  ax.plot(ranking['Year of Rank'] , ranking['Rank'], linewidth=2, marker='o')
  ax.set_xticks(ranking['Year of Rank'].values)
  ax.set_yticks(ranking['Rank'].values)
  ax.invert_yaxis()
  plt.title('{} Rankings'.format(coaster_name), fontsize=16)
  plt.xlabel('Year')
  plt.ylabel('Ranking')
  
  plt.show()
  plt.close()

# Create a plot of The voyage ranking over time
plot_ranking_overtime('The Voyage', 'Holiday World', golden_wood)
plt.clf()
```

```python
# 3 Creating a function to plot rankings over time for 2 roller coasters
def two_plot_ranking_overtime(coaster_name1, park_name1, coaster_name2, park_name2, ranking_df):
  ranking1_df = ranking_df[(ranking_df['Name'] == coaster_name1) & (ranking_df['Park'] == park_name1)]
  ranking2_df = ranking_df[(ranking_df['Name'] == coaster_name2) & (ranking_df['Park'] == park_name2)]

  fig, ax = plt.subplots(figsize=(8, 6))
  ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
  
  ax.plot(ranking1_df['Year of Rank'] , ranking1_df['Rank'], linewidth=2, color='red', label=coaster_name1, marker='o')
   
  ax.plot(ranking2_df['Year of Rank'], ranking2_df['Rank'], linewidth=2, color='green', label=coaster_name2, marker='o')
  
  plt.title('{} Ranking vs {} Ranking'.format(coaster_name1, coaster_name2), fontsize=14)
  plt.xlabel('Year')
  plt.ylabel('Ranking')
  ax.invert_yaxis()
  plt.legend()


  plt.show()
  plt.close()

# Create a plot of El Toro and Boulder Dash roller coasters
two_plot_ranking_overtime('El Toro','Six Flags Great Adventure','Boulder Dash','Lake Compounce', golden_wood)
```

```python
# 4 Creating a function to plot top n rankings over time
def n_rankings(n, ranking_df):
  n_ranking = ranking_df[ranking_df.Rank <= n]
  fig, ax = plt.subplots(figsize=(10, 8))

  for coaster in set(n_ranking['Name']):
    coaster_ranking = n_ranking[n_ranking['Name'] == coaster]
    ax.plot(coaster_ranking['Year of Rank'], coaster_ranking.Rank, label=coaster)
  
  ax.set_yticks([i for i in range(1, 6)])
  ax.invert_yaxis()
  
  plt.title('Top {} Roller Coasters Ranking'.format(n+5), fontsize=16)
  plt.xlabel('Year')
  plt.ylabel('Ranking')
  plt.legend(loc=4)

  plt.show()
  plt.close()
  
# Create a plot of top n rankings over time
n_rankings(5, golden_wood)
plt.clf
```

```python
# 5 loading roller coaster data and checking
roller_coasters = pd.read_csv('roller_coasters.csv')
print(roller_coasters.head())
```

```python
# 6 Creating a function to plot histogram of column values
def plot_hist(dataframe, column):
  bins = len(dataframe[column])

  fig, ax = plt.subplots(figsize=(8, 6))
  ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
  ax.set_facecolor("#ededed")
  plt.hist(dataframe[column].dropna(), edgecolor='black', color='#76bbff')
  plt.title('Roller Coasters Histogram {}'.format(column), fontsize=14)
  plt.xlabel(column)
  plt.ylabel('Count')
  plt.show()
  plt.close()
  print('\n')
  
# Creating histogram of roller coaster speed
plot_hist(roller_coasters, 'speed')
plt.clf()

# Creating histogram of roller coaster length
plot_hist(roller_coasters, 'length')
plt.clf()

# Creating histogram of roller coaster number of inversions
plot_hist(roller_coasters, 'num_inversions')
plt.clf()

# Creating a function to plot histogram of height values
plot_hist(roller_coasters, 'height')
plt.clf()

# Creating a histogram of roller coaster height
plot_hist(roller_coasters, 'length')
plt.clf()
```

```python
# 7 Creating a function to plot inversions by coaster at park
def plot_bar(dataframe, park_name):

    dataframe = dataframe[dataframe['park'] == park_name]
    dataframe = dataframe.sort_values('num_inversions', ascending=False)
    coaster_names = dataframe['name']
    number_inversions = dataframe['num_inversions']

    fig, ax = plt.subplots(figsize=(10, 8))
    ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
    ax.set_facecolor("#ededed")
    plt.bar(range(len(number_inversions)), number_inversions, edgecolor='black', color='#ffba76')
    ax.set_xticks(range(len(coaster_names)))
    ax.set_xticklabels(coaster_names, rotation=90)
    plt.title('Number of Inversions Per Coaster at {}'.format(park_name), fontsize=14)
    plt.xlabel('Roller Coaster')
    plt.ylabel('Counter of Inversions')
    plt.show()
    plt.close()


# Create barplot of inversions by roller coasters
plot_bar(roller_coasters, 'Six Flags Great Adventure')
```

```python
# 8 Creating a function to plot a pie chart of status.operating
def plot_pie(dataframe):
    
    operating_coaster = dataframe[dataframe.status == 'status.operating']
    num_operation_coaster = len(operating_coaster)

    closed_definitely_coaster = dataframe[dataframe.status == 'status.closed.definitely']
    num_closed_coaster = len(closed_definitely_coaster)

    status_counts = [num_operation_coaster, num_closed_coaster]

    plt.figure(figsize=(8, 6))
    plt.pie(status_counts, autopct='%1d%%', colors=['#76bbff', '#ff7677'])
    plt.axis('equal')
    plt.title('Status from every Roller Coaster', fontsize=16)
    plt.legend(labels=['Operating', 'Closed'], loc='right')
    plt.show()
    plt.close()

# Create pie chart of roller coasters
plot_pie(roller_coasters)
plt.clf()
```

```python
# 9 Creating a function to plot scatter of any two columns
def plot_scatter(column1, column2, dataframe):
    
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
    plt.scatter(dataframe[column1], dataframe[column2], edgecolors=['blue', 'red'])
    plt.title('Scatter plot of {} vs {}'.format(column1, column2), fontsize=16)
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()
    plt.close()

#creating a scatter plot speed vs height
plot_scatter('speed', 'height', roller_coasters)
plt.clf

#discovering the outliers
# print(roller_coasters[roller_coasters.height >= 150])

# Create a function to plot scatter of speed vs height
def plot_scatter_speed_height(dataframe):

    #taking out the outliers
    dataframe = dataframe[dataframe.height < 150]

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
    plt.scatter(dataframe['height'], dataframe['speed'], edgecolors=['blue', 'green'])
    plt.title('Scatter plot speed vs height', fontsize=16)
    plt.xlabel('Height')
    plt.ylabel('Speed')
    plt.show()
    plt.close()

# Create a scatter plot of roller coaster height by speed
plot_scatter_speed_height(roller_coasters)
plt.clf
```
